**Data tanpa outlier**

Mendeteksi dan mengeliminasi outlier dengan menggunakan metode **IQR** ***(Interquartile Range)*** yaitu metode yang berfungsi dengan menghitung rentang interkuartil dan kemudian menggunakan nilai tersebut untuk menentukan batas bawah dan batas atas yang mewakili rentang data yang dianggap normal. Nilai yang berada di luar batas ini dianggap sebagai outlier.

```python
# mengidentifikasi dan mengeliminasi outlier dari variabel "harga"

Q1 = dfJ["harga"].quantile(0.25)
Q2 = dfJ["harga"].quantile(0.50)
Q3 = dfJ["harga"].quantile(0.75)
IQR = Q3 - Q1 #IQR is interquartile range.

no_outlier = (dfJ["harga"] >= Q1 - 1.5 * IQR) & (dfJ["harga"] <= Q3 + 1.5 *IQR)
dfJno_out = dfJ[no_outlier]

dfJ_out = dfJ[~no_outlier]
```

```python
# visualisasi distribusi variabel "harga" setelah menghilangkan outlier
no_outlier_harga = sns.boxplot(x="harga", data = dfJno_out)
```
